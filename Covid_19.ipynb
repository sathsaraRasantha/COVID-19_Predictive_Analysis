```
import requests
import json
import pandas as pd

data=requests.get('https://www.hpb.health.gov.lk/api/get-current-statistical')
d=data.text
new=json.loads(d)

data1=[]
for i in range(0,35):
  data2=new['data']['hospital_data'][i]
  data1.append(data2)

data1
   


df = pd.json_normalize(data1)
df
```
