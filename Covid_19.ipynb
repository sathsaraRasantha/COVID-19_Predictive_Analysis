***Data Modeling***

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
```

```

colombo=pd.read_csv("colombo_full_count.csv")
kalutara=pd.read_csv("kalutara_full_count.csv")
gampaha=pd.read_csv("gampaha_full_count.csv")
other_areas=pd.read_csv("other_full_count.csv")
quarantine_centers=pd.read_csv("quarantine_full_count.csv")
```

Function to prepare data

```
def prepare_data(ts_data, n_features):
	X, y =[],[]
	for i in range(len(ts_data)):
		
		end_ix = i + n_features
		
		if end_ix > len(ts_data)-1:
			break
		
		seq_x, seq_y = ts_data[i:end_ix], ts_data[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return np.array(X), np.array(y)
```

```
n_steps=5
n_features=1

ts_colombo=prepare_data(colombo['Count'],n_steps)

X_colombo=ts_colombo[0]
y_colombo=ts_colombo[1]

X_colombo=X_colombo.reshape((X_colombo.shape[0],X_colombo.shape[1],n_features))
X_colombo.shape


```

LSTM model for COVID data

```
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, n_features)))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
tf.config.run_functions_eagerly(True)
model.fit(X_colombo, y_colombo, epochs=100, verbose=1)
```

Predictions for next 10 days ( COLOMBO District)

```
x_input_colombo = np.array(colombo['Count'][-5:])
temp_input_colombo=list(x_input_colombo)
lst_output_colombo=[]
i=0
while(i<10):
    
    if(len(temp_input_colombo)>5):
        x_input_colombo=np.array(temp_input_colombo[1:])
        print("{} day input {}".format(i,x_input_colombo))
        #print(x_input)
        x_input_colombo = x_input_colombo.reshape((1, n_steps, n_features))
        #print(x_input)
        yhat = model.predict(x_input_colombo, verbose=0)
        print("{} day output {}".format(i,yhat))
        temp_input_colombo.append(yhat[0][0])
        temp_input_colombo=temp_input_colombo[1:]
        #print(temp_input)
        lst_output_colombo.append(yhat[0][0])
        i=i+1
    else:
        x_input_colombo = x_input_colombo.reshape((1, n_steps, n_features))
        yhat = model.predict(x_input_colombo, verbose=0)
        print(yhat[0])
        temp_input_colombo.append(yhat[0][0])
        lst_output_colombo.append(yhat[0][0])
        i=i+1
    

print(lst_output_colombo)
```
